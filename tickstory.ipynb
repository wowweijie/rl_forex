```python
from finrl.preprocessing import influx_data
from finrl.preprocessing import data
```

Export as csv from your influxdb. Please create currency pair folder. Under it create "raw" and "ohlc" folder.  Please use 3600 as the window period to represent a one hour window period. 

```python
influx_data.export_csv_from_influx_chicago('NZDUSD', 3600, 'chicago_pmi/NZDUSD/raw')
```

Your csv is now inside "raw". Load it as a pandas dataframe

```python
raw_df = data.load_dataset('chicago_pmi/NZDUSD/raw/NZDUSD_Chicago_Pmi_2018-01-31.csv')
raw_df
```

Format it to OHLC

```python
ohlc_df = data.format_ohlc(raw_df)
ohlc_df
```

Export as CSV to "ohlc" folder

```python
data.export_dataset(ohlc_df, 'chicago_pmi/NZDUSD/ohlc/NZDUSD_Chicago_Pmi_2018-01-31.csv')
```
